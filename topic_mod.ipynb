```python
from nltk import word_tokenize,pos_tag
```

```python
def nouns_adj(text):
    is_noun_adj=lambda pos: pos[:2]=='NN' or pos[:2]=='JJ'
    tokens=word_tokenize(text)
    noun_adj=[x for (x,pos) in pos_tag(tokens) if is_noun_adj(pos)]
    return ' '.join(noun_adj)
```

```python
sent='Ayesha is a very good woman. She is a resident of Jharkhand. She cooks yummy pancakes. Her friends also call her aasha.'
```

```python
nouns_adj(sent)
```

```python
temp2_clean=pd.DataFrame(temp2.ABSTRACT.apply(nouns_adj))
temp2_clean
```

```python
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import numpy as np
```

```python
mat=CountVectorizer(max_df=0.8)
temp2_mat=mat.fit_transform(temp2_clean.ABSTRACT)
temp2_matn=pd.DataFrame(temp2_mat.toarray(),columns=mat.get_feature_names())
temp2_matn.index=temp2_clean.index
temp2_matn
```

```python
corpusmat=matutils.Sparse2Corpus(scipy.sparse.csr_matrix(temp2_matn.T, dtype=float))
id2words=dict((i,j) for j,i in mat.vocabulary_.items())
```
