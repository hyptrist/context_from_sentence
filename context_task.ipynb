```python
def nouns(text):                                                             
    is_noun=lambda pos: pos[:2]=='NN'
    tokens=word_tokenize(text)
    noun=[x for (x,pos) in pos_tag(tokens) if is_noun(pos)]
    return ' '.join(noun)
```
